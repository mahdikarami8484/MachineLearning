```python
import pandas as pd
```

```python
df = pd.read_csv("ExamScore.csv")
```

```python
import numpy as np
```

```python
hours = np.array(df["StudyHours"]).reshape(-1, 1)
hours
```

```python
scores = np.array(df['ExamScore']).reshape(-1, 1)
scores
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor(max_depth=10)
model.fit(hours, scores)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(hours, scores, "o")
plt.plot(hours, model.predict(hours))
plt.show()
```
