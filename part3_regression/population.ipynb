```python
import pandas as pd
```

```python
df = pd.read_csv("population.csv")
```

```python
import numpy as np
```

```python
year = np.array(df["Year"]).reshape(-1, 1)
year
```

```python
population = np.array(df["Population"]).reshape(-1, 1)
population
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(year, population, "*")
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(year, population)
```

```python
m = model.coef_[0]
b = model.intercept_

print(m, b)
```

```python
plt.plot(year, population, "o")
plt.plot(year, model.predict(year))
plt.show()
```

```python
population_pr = model.predict([[2050]])
print("{:,.0f}".format(population_pr[0,0]))
```
