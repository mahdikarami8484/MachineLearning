```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [2, 6, 7, 1]

plt.plot(x, y, marker='*',color="green")
plt.title("Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x = ["red", "green", "blue", "orange"]
y = [5, 20, 18, 10]

plt.bar(x, y, color="red")
plt.title("Example")
plt.xlabel("Color")
plt.ylabel("Count")
plt.show()
```

```python
x = ["red", "green", "blue", "orange"]
y = [5, 20, 18, 10]

plt.pie(y, labels=x, autopct="%1.1f%%", colors=x)
plt.title("test")
plt.show()
```

```python
import numpy as np

x = np.random.randint(0, 10, (1000, 1))

plt.hist(x, bins=10)
plt.show()
```

```python
x = np.random.rand(64, 64)

plt.imshow(x, cmap="hot")
plt.colorbar()
plt.show()
```
